```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread, imshow, imsave
from skimage import data_dir, util, exposure
from skimage.transform import radon, rescale

image_rgb = imread('/Users/wajdiahmed/Kaggle/TGA_RGB_90x90.png')
image_gray = imread('/Users/wajdiahmed/Kaggle/TGA_GRAY_90x90.png')

print(image_rgb.shape)
print(image_gray.shape)

print(image_gray)


imshow(image_gray)

plt.show()

imshow(image_rgb)

plt.show()


a = np.zeros((90, 90, 3))
b = image_gray
c = image_rgb


b= np.reshape(b,newshape=(90,90,1))
print(b.shape)


p2, p98 = np.percentile(b[:,:,0], (40, 60))



a[:,:,2] = exposure.rescale_intensity(b[:,:,0], in_range=(p2, p98))
a[:,:,1] = exposure.equalize_hist(b[:,:,0])
a[:,:,0] = exposure.equalize_adapthist(b[:,:,0], clip_limit=0.3)



imshow(a)


plt.show()


print(a.shape)


```
